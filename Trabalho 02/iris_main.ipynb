# Coletando resultados

A seguir temos a tabela com média e desvio padrão das taxas de acerto de cada
classificador

```python
data = {
    'Média': [np.mean(knn_one_neighbor), np.mean(knn_three_neighbor), np.mean(knn_five_neighbor), np.mean(knn_seven_neighbor), np.mean(dmc)],
    'Desvio Padrão': [np.std(knn_one_neighbor), np.std(knn_three_neighbor), np.std(knn_five_neighbor), np.std(knn_seven_neighbor), np.std(dmc)]
}
line_names = ['KNN - K=1', 'KNN - K=3', 'KNN - K=5', 'KNN - K=7', 'DMC']
pd.DataFrame(data, index=line_names)
```
